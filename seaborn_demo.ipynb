```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.boxplot(data=tips,y="total_bill",x="day", hue="smoker")
```

```python
sns.catplot(data=tips,y="total_bill",x="time",kind="bar")
# tips.groupby("time").mean()
```

## Relational plots

```python
sns.scatterplot(x="total_bill",y="tip", hue="sex", style="time", data = tips)
```

```python
tips [ tips['size'] == 1]
```

```python
sns.relplot(x="total_bill",y="tip", hue="sex", style="smoker", col="time", data = tips)
```

## regression plot

```python
sns.regplot(x="total_bill",y="tip", data = tips)
```

```python
sns.lmplot(x="total_bill",y="tip", data = tips,col="time")
```
